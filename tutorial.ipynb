```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
from RanDepict import RandomDepictor
```

```python
smiles = 'CNC(=O)C1=NC=CC(OC2=CC=C(NC(=O)NC3=CC(=C(Cl)C=C3)C(F)(F)F)C=C2)=C1'
```

```python
n_cols, n_rows = 5, 2

data = []
with RandomDepictor(hand_drawn=True) as depictor:
    for i in range(n_cols * n_rows):
        data.append(depictor(smiles))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(2*n_cols,2*n_rows))
for i, arr in enumerate(data):
    ax = axes[i//n_cols, i%n_cols]
    ax.imshow(Image.fromarray(arr))
    ax.axis('off')

plt.show()
```
